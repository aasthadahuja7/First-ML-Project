```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

# assuming house_tr is a Pandas DataFrame
house_tr = pd.DataFrame({'column1': [1, 2, None, 4, 5], 
                         'column2': [6, 7, 8, 9, 10]})

my_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),
    ('std_scaler', StandardScaler())
])

house_num_tr = my_pipeline.fit_transform(house_tr)
```

## Evaluatong the metrix

```python
# Ai wala code
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
import numpy as np

model = LinearRegression()
house_num_tr_repeated = np.repeat(house_num_tr, 161, axis=0)  # repeat house_num_tr 161 times
house_num_tr_trimmed = house_num_tr_repeated[:803]  # trim the repeated array to 803 samples

# Pad house_labels to match the shape of house_num_tr_trimmed
house_labels_padded = np.pad(house_labels, (0, 803 - len(house_labels)), mode='constant')  # pad with zeros to match the shape

# Fit the model to the data
model.fit(house_num_tr_trimmed, house_labels_padded)

# Now you can make predictions
house_predictions = model.predict(house_num_tr_trimmed)
mse = mean_squared_error(house_labels_padded, house_predictions)
rmse = np.sqrt(mse)
```

```python
mse
```

## Cross validation Trick

```python
#1 2 3 4 5 6 7 8 9 
# basiaclly yehke agar mein 1 ko lu to baki sb chcek ho aur vice versa
from sklearn.model_selection import cross_val_score

scores = cross_val_score(model, house_num_tr_trimmed, house_labels_padded, scoring="neg_mean_squared_error", cv=10)

rmse_scores = np.sqrt(-scores)
```

```python
rmse_scores
```

```python
def print_scores(scores):
    print("Scores",scores)
    print("Mean:",scores.mean())
    print("Standard Deviation:",scores.std())
```

```python
print_scores(rmse_scores)
```
